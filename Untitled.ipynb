```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import optimizers, Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation, Dense, Flatten, Dropout
import os
import numpy as np
import matplotlib.pyplot as plt
import imageio
import pickle
```

```python
final_tr_data = []
final_tr_label = []
final_te_data = []
final_te_label = []

def unpickle(file):
  fo = open(file, 'rb')
  dict = pickle.load(fo, encoding='latin1')
  fo.close()
  return dict

for i in range(1, 6):
  dataName = 'cifar-10-batches-py//data_batch_' + str(i)
  xtr = unpickle(dataName)
  print(dataName + " is loading...")
  for j in range(0, 10000):
    img = np.reshape(xtr['data'][j], (3,32,32)).transpose(1,2,0)
    picName = 'train//' + str(xtr['labels'][j]) + '_' + str(j+(i-1)*10000) + '.jpg'
    if not os.path.exists('./train'):
      os.mkdir('./train')
    imageio.imwrite(picName, img)
    final_tr_data.append(img)
    final_tr_label.append(xtr['labels'][j])
  print(dataName + " loaded")

print("Test batch loading...")
testXtr = unpickle('cifar-10-batches-py//test_batch')
for i in range(0, 10000):
  img = np.reshape(testXtr['data'][i], (3,32,32)).transpose(1,2,0)
  picName = 'test//' + str(testXtr['labels'][i]) + '_' + str(i) + '.jpg'
  if not os.path.exists('./test'):
    os.mkdir('./test')
  imageio.imwrite(picName, img)
  final_te_data.append(img)
  final_te_label.append(testXtr['labels'][i])
print('Test batch loaded')
```

```python
x_train = np.array(final_tr_data)
print(x_train.shape)
y_train = np.array(final_tr_label).reshape(50000, 1)
print(y_train.shape)

x_test = np.array(final_te_data)
print(x_test.shape)
y_test = np.array(final_te_label).reshape(10000, 1)
print(y_test.shape)
```

```python
category_dict = {0:'airplane',1:'automobile',2:'bird',3:'cat',4:'deer',5:'dog', 6:'frog',7:'horse',8:'ship',9:'truck'}
plt.figure()
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(x_train[i])
  plt.ylabel(category_dict[final_tr_label[i]])
plt.show()
```

```python
# Normalize pixels
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

num_classes = 10
```

```python
def CNN_classification_model(input_size=x_train.shape[1:]):
  model = Sequential()
  # First Block
  model.add(Conv2D(32, (3,3), padding='same', input_shape=input_size))
  model.add(Activation('relu'))
  model.add(Conv2D(32, (3,3)))
  model.add(Activation('relu'))
  model.add(MaxPooling2D(pool_size=(2,2), strides=1))
  # Second block
  model.add(Conv2D(64, (3,3)))
  model.add(Activation('relu'))
  model.add(Conv2D(64, (3,3)))
  model.add(Activation('relu'))
  model.add(MaxPooling2D(pool_size=(2,2)))
  # Flatten
  model.add(Flatten())
  # Fully connected layer
  model.add(Dense(128))
  model.add(Activation('relu'))
  model.add(Dropout(.25))
  model.add(Dense(num_classes))
  model.add(Activation('softmax'))
  # Compile
  opt = optimizers.Adam(learning_rate=0.0001)
  model.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
  return model
model = CNN_classification_model()
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
model_name = 'final_cifar2.h5'
model_check = ModelCheckpoint(model_name, monitor='loss', verbose=1, save_best_only=True)

trained_weights_path = 'cifar10_weights.h5'
if os.path.exists(trained_weights_path):
  model.load_weights(trained_weights_path, by_name=True)
# Train
model.fit(x_train, y_train, batch_size=32, epochs=10, verbose=1, callbacks=[model_check])
```

```python
new_model = CNN_classification_model()
new_model.load_weights(model_name)

model.evaluate(x_test, y_test, verbose=1)
```
